```python
from weiboPredict import framework as fw
```

```python
fw.loadData()
```

```python
weibo_train_data_u_sum=fw.weibo_train_data.groupby('uid').sum()
```

```python
weibo_train_data_u_count=fw.weibo_train_data.groupby('uid').mid.count()
```

```python
weibo_train_data_u = weibo_train_data_u_sum
```

```python
weibo_train_data_u_sum.index.name
```

```python
weibo_train_data_u['weibo_count'] = weibo_train_data_u_count
```

```python
weibo_train_data_u.head()
```

```python
weibo_train_data_u.weibo_count.isnull().value_counts()
```

```python
weibo_train_data_u['forward_avg'] = weibo_train_data_u.forward_count/weibo_train_data_u.weibo_count
```

```python
weibo_train_data_u.head()
```

```python
from pandas import DataFrame, Series
```

```python
weibo_train_data_u.forward_avg=weibo_train_data_u.forward_avg.round()
```

```python
weibo_train_data_u.head()
```

```python
weibo_train_data_u['comment_avg'] = weibo_train_data_u.comment_count/weibo_train_data_u.weibo_count
```

```python
weibo_train_data_u.comment_avg=weibo_train_data_u.comment_avg.round()
```

```python
weibo_train_data_u['like_avg'] = weibo_train_data_u.like_count/weibo_train_data_u.weibo_count
```

```python
weibo_train_data_u.like_avg=weibo_train_data_u.like_avg.round()
```

```python
weibo_train_data_u.head()
```

```python
weibo_train_data_u.sort('forward_avg',ascending=False)
```

```python
fw.weibo_predict_data.set_index('uid',inplace=True)
```

```python
fw.weibo_predict_data[['forward_count','comment_count','like_count']] = weibo_train_data_u[['forward_avg','comment_avg','like_avg']]
```

```python
fw.weibo_predict_data.head(20)
```

```python
fw.weibo_predict_data.forward_count.isnull().value_counts()
```

```python
fw.weibo_predict_data.comment_count.isnull().value_counts()
```

```python
fw.weibo_predict_data.like_count.isnull().value_counts()
```

```python
fw.weibo_predict_data.forward_count.describe()
```

```python
fw.weibo_predict_data.comment_count.describe()
```

```python
fw.weibo_predict_data.like_count.describe()
```

```python
fw.weibo_predict_data.forward_count.fillna(fw.weibo_predict_data.forward_count.mean().round(),inplace=True)
```

```python
fw.weibo_predict_data.comment_count.fillna(fw.weibo_predict_data.comment_count.mean().round(),inplace=True)
```

```python
fw.weibo_predict_data.like_count.fillna(fw.weibo_predict_data.like_count.mean().round(),inplace=True)
```

```python
fw.weibo_predict_data.forward_count.isnull().value_counts()
```

```python
fw.weibo_predict_data[['mid','forward_count','comment_count','like_count']].to_csv('result/result_u_avg.csv',sep=',',float_format='%d')
```
